```python
**Dicionário de Variáveis:**
* **crim:** Taxa de crime per capita por cidade.
* **zn:** Proporção de terrenos residenciais zoneados para lotes grandes.
* **indus:** Proporção de acres de negócios não varejistas.
* **chas:** Variável dummy do Rio Charles (1 se faz fronteira com o rio; 0 caso contrário).
* **nox:** Concentração de óxidos de nitrogênio.
* **rm:** Número médio de quartos por habitação.
* **age:** Proporção de unidades construídas antes de 1940.
* **dis:** Distâncias ponderadas para centros de emprego em Boston.
* **rad:** Índice de acessibilidade a rodovias.
* **tax:** Taxa de imposto sobre a propriedade.
* **ptratio:** Razão aluno-professor por cidade.
* **b:** Proporção de população negra por cidade.
* **lstat:** Porcentagem de status inferior da população (indicador socioeconômico).
* **medv:** Valor mediano das casas ocupadas pelos proprietários (em $1000s).
```
